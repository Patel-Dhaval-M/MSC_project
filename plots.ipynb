```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams["figure.figsize"] = [6,5]
plt.grid(True, which='both')
plt.plot(np.linspace(-3, 3, 100), np.tanh(np.linspace(-3, 3, 100)))
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.title("Tanh")
plt.xlabel("Input")
plt.ylabel("Output")
```

```python
df1 = pd.read_csv("ddpg/log-files/Hopper-v1/no-of-update_1/1/log.csv")
df5 = pd.read_csv("ddpg/log-files/Hopper-v1/no-of-update_5/1/log.csv")
df10 = pd.read_csv("ddpg/log-files/Hopper-v1/no-of-update_10/1/log.csv")
df20 = pd.read_csv("ddpg/log-files/Hopper-v1/no-of-update_20/1/log.csv")
df20.head()
```

```python
plt.rcParams["figure.figsize"] = [9,4]
no_of_updates_1, = plt.plot(df1["Total_Episode"], df1["Target_Reward"])
no_of_updates_5, = plt.plot(df5["Total_Episode"], df5["Target_Reward"])
no_of_updates_10, = plt.plot(df10["Total_Episode"], df10["Target_Reward"])
no_of_updates_20, = plt.plot(df20["Total_Episode"], df20["Target_Reward"])
plt.xlabel("Number of episodes")
plt.ylabel("Rewards")
plt.title("Hopper - DDPG-R")
plt.legend([no_of_updates_1, no_of_updates_5, no_of_updates_10, no_of_updates_20], ["1 update", "5 updates", "10 updates", "20 updates"])
```
